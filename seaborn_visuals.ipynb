```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(x='sex',hue='day', data=tips,jitter=True)
plt.title('Strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
# Swarm plot
plt.figure(figsize=(15,15))
sns.swarmplot(x='day',y='tip',data=tips,facecolor='black')
sns.violinplot(x='day',y='tip',data=tips,color='black')
plt.show()
```

```python
#plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.boxplot(x='day', y='tip', data=tips,)
sns.violinplot(x='day', y='tip',data=tips,color='lightgrey')

plt.ylabel('tip ($)')
plt.show()
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
# subplot
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#jointplot
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
#density plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```

```python
#pairplot
sns.pairplot(tips)
plt.show()
```
