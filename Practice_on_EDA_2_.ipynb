###Label Encoder



```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/iris.csv')
```

```python
data.head()
```

```python
# changes does not effect on "data" dataframe
data1=data.copy()
```

```python
labelencoder =LabelEncoder()
data1.iloc[:,-1]=labelencoder.fit_transform(data.iloc[:,-1])
```

```python
data1.iloc[:,-1]
```

##One Hot Encoder


```python
#using sklearn 
from sklearn.preprocessing import OneHotEncoder
```

```python
data2= pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/iris.csv')
```

```python
data2
```

```python
# ceating instance of one hot encoder 
OHE= OneHotEncoder(handle_unknown='ignore')
```

```python
# Passing Bridge_Type_cat_Column
enc_df=pd.DataFrame(OHE.fit_transform(data2[['Species']]).toarray())
```

```python
OHE.fit_transform(data2[['Species']]).toarray()
```

```python
enc_df
```

```python
# merge with main df
data_final=data2.iloc[:,0:4].join(enc_df)
data_final
```
